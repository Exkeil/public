Опишите словами нулевую и альтернативную гипотезы. Проверьте гипотезы с помощью статистического критерия. Интерпретируйте получившиеся результаты. Уровень значимости = 5%.

Нулевая гипотеза: Различия между группами X и Y в оценках за тест нет__

Альтернативная гипотез По методике X студенты набрали больший балл, чем студенты, обучающиеся по методики Y__

```python
import pandas as pd
df = pd.read_csv("data/hw1.csv")
df.head
from scipy import stats
df['X'] = pd.to_numeric(df['X'], errors='coerce')
df['Y'] = pd.to_numeric(df['Y'], errors='coerce')
print(stats.shapiro(df['X']),stats.shapiro(df['Y']))
```

```python
stats.levene(df['X'], df['Y'])
stats.ttest_ind(df['X'], df['Y'], equal_var = False)
```

```python
# после результата теста т-критерия уэлча у нас нет достаточно оснований утверждать, что среднее значение статистически значимо различаются(прошу прощения, что не все результаты выводятся на отдельной строке, это было долго)  
```

Нулевая гипотеза:Прием добавки не приводит к статистически значимому уменьшению веса у тестовой группы за две недели.

Альтернативная гипотеза: Прием добавки приводит к статистически значимому уменьшению веса у тестовой группы за две недели.

```python
df2 = pd.read_csv('data/hw2.csv')
df2.head()
df2['before'] = pd.to_numeric(df2['before'], errors='coerce')
df2['after'] = pd.to_numeric(df2['after'], errors='coerce')
print(stats.shapiro(df2['before']),
stats.shapiro(df2['after']))
```

```python
stats.ttest_rel(df2['before'], df2['after'], alternative='less') # как я понимаю, у нас нет значимых доказательств отвергать нулевую гипотезу, то есть не факт, что добавка уменьшает или увеличивает вес
```
